```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame({'X':np.random.normal(0,2,1000),
                  'Y':np.random.normal(3,5,1000),
                  'Z':np.random.normal(4,10,1000)})
df.head()
```

```python
plt.style.use('seaborn')
sns.set()
fig = plt.figure(figsize=(4,3))
for col in df.columns:
    plt.hist(df[col],density=True,alpha=0.5,bins=30)
    sns.kdeplot(df[col],shade=True)
```

```python
plt.figure(figsize=(4,3))
for col in df.columns:
    sns.distplot(df[col])
```

```python
plt.figure(figsize=(4,3))
with sns.axes_style('white'):
    sns.jointplot('X','Z',df,kind='kde')

plt.figure(figsize=(4,3))
with sns.axes_style('white'):
    sns.jointplot('X','Y',df,kind='kde')
    
plt.figure(figsize=(4,3))
with sns.axes_style('white'):
    sns.jointplot('Z','Y',df,kind='kde')
```

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris,hue='species')# we can call this for the classes and hue is the categories
```

```python
# seaborn Relplot(relationship plot)
iris.head()
```

```python
# for having a different styles we can use relplot
sns.relplot(x='sepal_length',y='sepal_width',data=iris,
            hue='species',style='species')
```

```python
#relplot kind line
sns.relplot(x='sepal_length',y='sepal_width',dashes=False,markers=True,
            kind='line',data=iris,ci=None,hue='species',style='species')#ci=None is for clearing the shape
```

```python
#relplot facets
tips = sns.load_dataset('tips')
tips
```

```python
#making a relationship between bill and tip and smoker is category:
tips.time.unique()
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',
            style='smoker',col='time',data=tips)
```

```python
fmri = sns.load_dataset('fmri') 
fmri
```

```python
len(fmri.subject.unique())
```

```python
# see the relationship between the time point and the signal:
sns.relplot(x='timepoint',y='signal',hue='event',
           style='event',col='subject',col_wrap=3,#the numbers greater than 3 give the smaller shape and on the contrary
           kind='line',data=fmri.query("region=='frontal'"))#picking the data by query
```

```python
#catplot
sns.set(style='ticks',color_codes=True)
```

```python
sns.catplot(x='day',y='total_bill',data=tips,jitter=False)#jitter makes the graph more linear
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='box',hue='sex') #kind=box
```
