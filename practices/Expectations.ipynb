```python
#Law of large numbers
### Bernoulli:
# E[X] = p
p = 0.812
sampleMeanX = []
for N in range(1,10000,100):#sample
    sampleMeanX.append((np.random.rand(N)<=p).mean())
y= np.array(sampleMeanX)
x = np.arange(y.size)
plt.plot(x,y,linestyle='--')
```

```python
#Geometric(p), E[X] = 1/p
p = 0.312
print(1/p)
sampleMeanX = []
for N in range(1,10000,100):#sample
    sampleMeanX.append((np.random.geometric(p,N).mean()))
y= np.array(sampleMeanX)
x = np.arange(y.size)
plt.plot(x,y,linestyle='--')
```

```python
#binomial(n,p), E[X] = np
n = 350 #total of bernoulli trails
p = 0.312
print(n*p)
sampleMeanX = []
for N in range(1,10000,100):#sample
    sampleMeanX.append(((np.random.binomial(n,p,N)).mean()))
y= np.array(sampleMeanX)
x = np.arange(y.size)
plt.plot(x,y,linestyle='--')
```

```python
#(Gaussian)Normal(mu,sgma), E[X] = mu

mu = 2.4
sgma = 3.1
sampleMeanX = []

for N in range(1,10000,100):#sample
    sampleMeanX.append((np.random.normal(mu,sgma,N).mean()))
y = np.array(sampleMeanX)
x = np.arange(y.size)
plt.plot(x,y,linestyle='--')
```
