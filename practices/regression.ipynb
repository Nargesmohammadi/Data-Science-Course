```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 9)
A = np. array([x, np.ones(9)])
```

```python
y = [19, 20, 20.0, 21.5, 22, 23, 23, 25.5, 24]
```

```python
w = np.linalg.lstsq(A.T, y)[0]
```

```python
line = w[0]*x + w[1]
```

```python
plt.plot(x, line, 'r-')
```

```python
plt.plot(x, y, 'o')
```

```python
plt.show()
```
