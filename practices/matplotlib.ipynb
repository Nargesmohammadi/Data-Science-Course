```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-whitegrid')
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,1000)
y = np.sin(x)
ax.plot(x,y)
```

```python
plt.style.use('seaborn')
y1 = np.sin(x) #blue
y2 = np.cos(x) #green
y3 = y1+y2-0.5 #red
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
```

```python
#another way:
fig = plt.figure()
ax = plt.axes()
ax.plot(x,y1)
ax.plot(x,y2)
ax.plot(x,y3)
```
