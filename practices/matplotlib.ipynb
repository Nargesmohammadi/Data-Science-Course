```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-whitegrid')
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,1000)
y = np.sin(x)
ax.plot(x,y)
```

```python
plt.style.use('seaborn')
y1 = np.sin(x) #blue
y2 = np.cos(x) #green
y3 = y1+y2-0.5 #red
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
```

```python
#another way:
fig = plt.figure(figsize=(4,3))# 4 is the height and 3 is the width
ax = plt.axes()
ax.plot(x,y1)
ax.plot(x,y2)
ax.plot(x,y3)
```

```python
fig = plt.figure(figsize=(4,2))
ax = plt.axes()
ax.plot(x,y1,color='blue',linestyle='solid')
ax.plot(x,y2,color='g',linestyle='dashed')
ax.plot(x,y3,color=(0.3,0.6,0.7),linestyle='dashdot')
```

```python
#another way:
fig = plt.figure(figsize=(4,2))
ax = plt.axes()
ax.plot(x,y1,'-b',linewidth=3)
ax.plot(x,y2,'--g')
ax.plot(x,y3,':r')
```

```python
# axis limits:
plt.figure(figsize=(4,2),dpi=100) #dpi is for making image crips
plt.plot(x,y1,'-.r',linewidth=0.8)
plt.plot(x,y2,':b')
plt.plot(x,y3,'-k')
# plt.xlim(-2,12)
# plt.ylim(-2,1.5)
#plt.axis([-2,12,-2,1.5])
plt.axis('tight')
```

```python
plt.figure(figsize=(4,2),dpi=100)
plt.plot(x,2*x+3,'--k')
plt.xlim(10,0)#reverse
```

```python
# labels and legends
plt.figure(figsize=(4,3),dpi=100) #dpi is for making image crips
plt.plot(x,y1,'-.r',label='sin(x)')
plt.plot(x,y2,':b',label='cos(x)')
plt.plot(x,y3,'-k',label='sin(x)+cos(x)-0.5')
plt.axis('equal')
plt.legend(loc='lower left')
plt.title("Sin-Cos-Curves")
plt.xlabel("X values")
plt.ylabel("Y values")
```

```python
#pyplot vs axes
plt.figure(figsize=(4,3),dpi=100) #dpi is for making image crips
ax = plt.axes()
ax.plot(x,y1,'-.r',label='sin(x)')
ax.plot(x,y2,':b',label='cos(x)')
ax.plot(x,y3,'-k',label='sin(x)+cos(x)-0.5')
ax.set(xlim=(-1,11),ylim=(-4,3),xlabel="X values",
       ylabel="Y values",title='Sin-Cos-curves')
#plt.axis('equal')
plt.legend(loc='lower left')
```

```python
# markers
markers = ['o','.','x','+','v','^','<','>','s','d']
x = np.linspace(-np.pi,np.pi,30)
y1 = np.sin(x)
plt.figure(figsize=(4,3),dpi=100)
plt.plot(x,y1,markers[8],color='black',markersize=5)
```
