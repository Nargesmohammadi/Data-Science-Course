```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-whitegrid')
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,1000)
y = np.sin(x)
ax.plot(x,y)
```

```python
plt.style.use('seaborn')
y1 = np.sin(x) #blue
y2 = np.cos(x) #green
y3 = y1+y2-0.5 #red
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
```

```python
#another way:
fig = plt.figure(figsize=(4,3))# 4 is the height and 3 is the width
ax = plt.axes()
ax.plot(x,y1)
ax.plot(x,y2)
ax.plot(x,y3)
```

```python
fig = plt.figure(figsize=(4,2))
ax = plt.axes()
ax.plot(x,y1,color='blue',linestyle='solid')
ax.plot(x,y2,color='g',linestyle='dashed')
ax.plot(x,y3,color=(0.3,0.6,0.7),linestyle='dashdot')
```

```python
#another way:
fig = plt.figure(figsize=(4,2))
ax = plt.axes()
ax.plot(x,y1,'-b',linewidth=3)
ax.plot(x,y2,'--g')
ax.plot(x,y3,':r')
```

```python
# axis limits:
plt.figure(figsize=(4,2),dpi=100) #dpi is for making image crips
plt.plot(x,y1,'-.r',linewidth=0.8)
plt.plot(x,y2,':b')
plt.plot(x,y3,'-k')
# plt.xlim(-2,12)
# plt.ylim(-2,1.5)
#plt.axis([-2,12,-2,1.5])
plt.axis('tight')
```

```python
plt.figure(figsize=(4,2),dpi=100)
plt.plot(x,2*x+3,'--k')
plt.xlim(10,0)#reverse
```

```python
# labels and legends
plt.figure(figsize=(4,3),dpi=100) #dpi is for making image crips
plt.plot(x,y1,'-.r',label='sin(x)')
plt.plot(x,y2,':b',label='cos(x)')
plt.plot(x,y3,'-k',label='sin(x)+cos(x)-0.5')
plt.axis('equal')
plt.legend(loc='lower left')
plt.title("Sin-Cos-Curves")
plt.xlabel("X values")
plt.ylabel("Y values")
```

```python
#pyplot vs axes
plt.figure(figsize=(4,3),dpi=100) #dpi is for making image crips
ax = plt.axes()
ax.plot(x,y1,'-.r',label='sin(x)')
ax.plot(x,y2,':b',label='cos(x)')
ax.plot(x,y3,'-k',label='sin(x)+cos(x)-0.5')
ax.set(xlim=(-1,11),ylim=(-4,3),xlabel="X values",
       ylabel="Y values",title='Sin-Cos-curves')
#plt.axis('equal')
plt.legend(loc='lower left')
```

```python
# markers
markers = ['o','.','x','+','v','^','<','>','s','d']# 'x' and '+' are not working at seaborn style
x = np.linspace(-np.pi,np.pi,30)
y1 = np.sin(x)
plt.figure(figsize=(4,3),dpi=100)
plt.plot(x,y1,markers[8],color='black',markersize=5,linestyle=':')
```

```python
plt.style.use('classic')
markers = ['o','.','x','+','v','^','<','>','s','d']
plt.figure(figsize=(4,3),dpi=100)
for m in markers:
    plt.plot(np.random.rand(5),np.random.rand(5),m,markersize=5)
```

```python
# scatter plot
plt.style.use('seaborn')
plt.figure(figsize=(4,3))
x = np.random.randn(100)
y = np.random.rand(100)
colors = np.random.rand(100)
sizes = 150*np.random.rand(100)# size between 0 and 150
plt.scatter(x,y,c=colors,s=sizes,alpha=0.5,cmap=plt.cm.ocean_r)
plt.colorbar()
```

```python
X,Y = np.meshgrid(np.linspace(0,5,50),np.linspace(0,5,40))
Z = 2*X-5*np.sin(Y)+2 #somefunction
plt.contour(X,Y,Z,20,cmap=plt.cm.Blues_r)# 20 is the number of the colorful curves
plt.colorbar()
```

```python
X,Y = np.meshgrid(np.linspace(0,5,50),np.linspace(0,5,40))
Z = 2*X-5*np.sin(Y)+2 #somefunction
plt.contourf(X,Y,Z,20,cmap=plt.cm.Blues_r) #contourf: f is fiiling between contours
plt.colorbar()
```

```python
#histograms
plt.figure(figsize=(4,3))
d1 = np.random.normal(0,0.7,1000)
d2 = np.random.normal(-5,2,1000)
d3 = np.random.normal(5,3,1000)
plt.hist(d1,alpha=0.5,bins=50,histtype='stepfilled')
plt.hist(d2,alpha=0.5,bins=50,histtype='stepfilled')
plt.hist(d3,alpha=0.5,bins=50,histtype='stepfilled')
```

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.head()
```

```python
plt.figure(figsize=(4,3))
parms = dict(histtype='stepfilled',alpha=0.4,bins=30)
plt.hist(iris.sepal_length,**parms)
plt.hist(iris.sepal_width,**parms)
plt.hist(iris.petal_length,**parms)
plt.hist(iris.petal_width,**parms)
```

```python
#subplot
fig,ax = plt.subplots(2,3,figsize=(4,3),sharex='col',sharey='row')#2 and 3 are an array, sharex and sharey are for removing the numbers of columns and rows
```

```python
fig,ax = plt.subplots(2,3,figsize=(4,3),sharex='col',sharey='row')
for i in range(2):
    for j in range(3):
        ax[i,j].plot(np.random.rand(10),np.random.rand(10),'sr')# s is the shape and r is the color
```

```python
fig,ax = plt.subplots(2,3,figsize=(4,3),sharex='col',sharey='row')
for i in range(2):
    for j in range(3):
        ax[i,j].plot(np.random.rand(10),np.random.rand(10),'sr')
        ax[i,j].set(xticks=[],yticks=[])#for having a very clean kind of input where all the ticks x and y
        
```

```python
fig,ax = plt.subplots(2,3,figsize=(4,3),sharex='col',sharey='row')
for i in range(2):
    for j in range(3):
        ax[i,j].imshow(np.random.rand(200,200),plt.cm.YlOrBr_r)#imshow is for showing the image
        ax[i,j].set(xticks=[],yticks=[])
```

```python
from sklearn.datasets import fetch_lfw_people
faces = fetch_lfw_people(min_faces_per_person=70)
plt.imshow(faces.images[4],plt.cm.bone)
```

```python
fig,ax = plt.subplots(4,5,figsize=(4,3),sharex='col',sharey='row')
k = 0
for i in range(4):
    for j in range(5):
        ax[i,j].imshow(faces.images[k],plt.cm.YlOrBr_r)#imshow is for showing the image
        ax[i,j].set(xticks=[],yticks=[])
        k+=1
```
