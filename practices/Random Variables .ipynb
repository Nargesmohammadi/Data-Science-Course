```python
import numpy as np
```

```python
#generate random values between 0 and 1
def BernulliTrail(p=0.5):
    X = int(np.random.rand()<=p) #generate n random variables
    return X
```

```python
print(BernulliTrail(0.7))
```

```python
counOnes = 0
countZeros = 0
p = 0.7 # the probablity os success
n = 10
for i in range(n):
    x = BernulliTrail(p)
    if x == 1:
        counOnes += 1
    else:
         countZeros += 1   
print('Success Fraction is: ', counOnes/n)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#return geometric random variables:
def GT(p):
    X = 1
    while True:
        if BernulliTrail(p) == 1:
            return X
        else:
            X += 1
```

```python
p = 0.2 #probability of success
print(GT(p)) 
```

```python
n = 1000
p = 0.6
G = np.zeros(n)
for i in range(n):
    G[i] = GT(p)
```

```python
sns.distplot(G)
```

```python
n = 1000
p = 0.2
G = np.zeros(n)
for i in range(n):
    G[i] = GT(p)

sns.distplot(G)
```

```python
n = 100000
p = 0.02
G = np.zeros(n)
for i in range(n):
    G[i] = GT(p)

plt.hist(G,density=True,bins=100)
sns.kdeplot(G,shade=True)
```

# Binomial

```python
def BT(n,p): # n is the total of bernoulli trails
    X = (np.random.rand(n)<=p).sum()
    return X
```

```python
print(BT(100,0.7))
```

```python
N = 10000 #the total of binomial experiments
n = 100 #the total of bernoulli trails
p = 0.612 
B = np.zeros(N) #binomial array
for i in range(N):
    B[i] = BT(n,p)
plt.hist(B,density=True,bins=20)
sns.kdeplot(B,shade=True)
```

# Exponential ditribution

```python
x = np.linspace(0,10,1000)
lmda = 1
fx = lmda*np.exp(-lmda*x)
plt.plot(x,fx)
```

```python
x = np.linspace(0,10,1000)
lmdaz = np.linspace(0.1,5,10)
for i in lmdaz:
    fx = i*np.exp(-i*x)
    plt.plot(x,fx,label=str(i))
plt.legend()
```

# Gaussian 

```python
x = np.linspace(-20,20,10000)
mu = 1
sgma = 1
fx = np.exp((-(x-mu)**2)/(2*(sgma**2)))/((2*np.pi*sgma**2)**0.5)
```

```python
plt.plot(x,fx)
```

```python
x = np.linspace(-20,20,10000)
muz = np.arange(1,20,2)
for mu in muz:
    sgma = 1
    fx = np.exp((-(x-mu)**2)/(2*(sgma**2)))/((2*np.pi*sgma**2)**0.5)
    plt.plot(x,fx)
```

```python
x = np.linspace(-20,20,10000)
muz = np.arange(-20,20,10)
for mu in muz:
    sgma = 1
    fx = np.exp((-(x-mu)**2)/(2*(sgma**2)))/((2*np.pi*sgma**2)**0.5)
    plt.plot(x,fx)
```

```python
x = np.linspace(-20,20,10000)
mu = 0
sgmaz = np.arange(1,10,2)
for sgma in sgmaz:
    fx = np.exp((-(x-mu)**2)/(2*(sgma**2)))/((2*np.pi*sgma**2)**0.5)
    plt.plot(x,fx,label=str(sgma))
plt.legend()
```
