```python
import numpy as np
```

```python
#generate random values between 0 and 1
def BernulliTrail(p=0.5):
    X = int(np.random.rand()<=p)
    return X
```

```python
print(BernulliTrail(0.7))
```

```python
counOnes = 0
countZeros = 0
p = 0.7
n = 10
for i in range(n):
    x = BernulliTrail(p)
    if x == 1:
        counOnes += 1
    else:
         countZeros += 1   
print('Success Fraction is: ', counOnes/n)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#return geometric random variables:
def GT(p):
    X = 1
    while True:
        if BernulliTrail(p) == 1:
            return X
        else:
            X += 1
```

```python
p = 0.2 #probability of success
print(GT(p)) 
```

```python
n = 1000
p = 0.6
G = np.zeros(n)
for i in range(n):
    G[i] = GT(p)
```

```python
sns.distplot(G)
```

```python
n = 1000
p = 0.2
G = np.zeros(n)
for i in range(n):
    G[i] = GT(p)

sns.distplot(G)
```

```python
n = 100000
p = 0.8
G = np.zeros(n)
for i in range(n):
    G[i] = GT(p)

plt.hist(G,density=True)
sns.kdeplot(G,shade=True)
```
