```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('seaborn-talk')
fig = plt.figure(figsize=(6,4))
x = np.linspace(0,20,1500)
y = np.cos(x)
plt.xlim(1,14) #set x limit of 1 to 14
# plt.ylim(-1,1.2) #set y limit of -1 to 1.2
plt.ylim(1.2,-1)#reverse the plot at y axis
plt.plot(x,y,color='orange',linestyle='dashed')
```
