```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
```

```python
df = pd.read_csv('/home/mitraa/Jupyter Notebook/covid-19_trendAnalysis/covid_19_clean_complete_2022.csv')
```

```python
m = folium.Map()
df2 = df[df['Date']==min(df['Date'])]
```

```python
df3 = df2.fillna(0)
df3
```

```python
for i in range(0,len(df3)):
    folium.Circle(location=[df3.iloc[i]['Lat'],df3.iloc[i]['Long']],
                 color='crimson',fill='crimson',
                 radius=int(df3.iloc[i]['Deaths'])).add_to(m)
m
```
